Q: [1541. Minimum Insertions to Balance a Parentheses String](https://leetcode.com/problems/minimum-insertions-to-balance-a-parentheses-string/)

**Key idea**:
* There are two kinds of additions that you can make:
	* `pairs`:
		* A *pair* of `)` for every fully orphaned `(`.
	* `ones`:
		* `ones_right`: A *single* `)` for every partially orphaned `(`, or, 
		* `ones_left`: A *single* `(` for every orphaned `)`.
* `2 * pairs + ones` is the desired minimum number of additions.



**Approach 1**:
* As you traverse through the string
	*  If `(`:
		*  Assume it is fully orphaned. 
		*  Increment `pairs`.
	* Else if `)`:
		* Check if next one is also `)`.
		* If yes: decrement `pairs` (our assumption that `(` is fully orphaned is wrong).
			* In code 1, the decrement happens later. So, `pairs >= 0`. It isn't necessary for the logic. Code 2 allows for it go < 0.
		* Else: increment `ones` (only one `)` needs to be added).
		* If `)` is orphaned (`pairs = 0`)
			* Increment `ones` (one `(` needs to be added before `)`)

```python
class Solution:
    def minInsertions(self, s: str) -> int:
        pairs, ones = 0, 0
        i, n = 0, len(s)
        
        while i < n:
            if s[i] == '(': pairs += 1                
            else:
                if i == n-1 or s[i+1] != ')': ones += 1
                else: i += 1                
                if not pairs: ones += 1
                else: pairs -= 1                
            i += 1
            
        return 2 * pairs + ones
```

**Approach 2**:
* The key idea is essentially the same. 
* `rights` corresponds to `pairs`. But here each `)` is treated individually.
* `ones` is split as described in Key Idea.
* Note that in a balanced string, we must have an even number of `)`.
* As you traverse through the string
	* If `(`:
		* Check if the *previous* `(` is partially orphaned, i.e., if `rights` is odd.
		* If yes: 
			* One `)` needs to be added.
			* Note that "partially orphaned" means one `)` was already found, which means we already decremented `rights` by 1 (see Else below). So, we need to decrement it by one more to remove the quota of 2 `rights` added to it when we initially scanned it (read the line below).
		* Add 2 `rights` for the *current* `(` (like Approach 1, assume it is fully orphaned).
	* Else (if `)`):
		* Decrement `rights` by 1
		* If a `)` is orphaned, i.e., `rights` < 0:
			* Add 1 `(` to its left
			* Add 1 `)` to its right (adding two `rights` here because you just decremented it by 1 above)

```python
class Solution:
    def minInsertions(self, s: str) -> int:
        rights, ones_right, ones_left = 0, 0, 0
        
        for c in s:
            if c == '(':
                if rights % 2:
                    rights -= 1
                    ones_right += 1
                rights += 2
            else:
                rights -= 1
                if rights < 0:
                    rights += 2
                    ones_left += 1                
                
        return rights + ones_right + ones_left
```
