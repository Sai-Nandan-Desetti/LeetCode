Q: [946. Validate Stack Sequences](https://leetcode.com/problems/validate-stack-sequences/)

* Use a stack to *simulate* the push-pop sequence.
* Go for a *greedy* approach: **Immediately pop any element that can be popped.**
* As you read an element from `pushed`, push it onto `stack`. As long as elements "can be popped" (`i` is used to determine this) from `stack`, pop them.
* When no more elements can be pushed/popped, check `stack`. If it's empty, then the simulated push-pop sequence is a valid one.

n: length of `pushed`/`popped`
* Time complexity: O(n)
* Space complexity: O(n)

To ponder:
* Does this code also work for non-unique elements?
(My playground for this question can be found [here](https://leetcode.com/playground/TTC59bsK).)

```python
class Solution:
    def validateStackSequences(self, pushed: List[int], popped: List[int]) -> bool:
        stack = []
        i = 0
        for x in pushed:
            stack.append(x)
            while stack and stack[-1] == popped[i]:
                i += 1
                stack.pop()
                
        return len(stack) == 0
```
