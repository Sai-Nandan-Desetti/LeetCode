Q: [918. Maximum Sum Circular Subarray](https://leetcode.com/problems/maximum-sum-circular-subarray/)

* Two cases (see [discussion](https://leetcode.com/problems/maximum-sum-circular-subarray/discuss/178422/One-Pass)):
  * The subarray is contained entirely within the body of the array.
  * The subarray is split across the head and the tail parts.
  * The desired maximum sum can be computed as follows:
\begin{equation}
  maxSum = \mathrm{max}(maxSum, array\_total - minSum),
\end{equation}
where $minSum$ is the minimum sum of a (non-circular) subarray in the given array.

* But there's a special case: Entire array has only negative numbers!
* Then, $minSum = array\_total$, thereby, $maxSum = 0$.
* Thus, in order to deal with the special case too, we return
\begin{equation}
  maxSum = \mathrm{max}(maxSum, array\_total - minSum) \hspace{2mm} \mathrm{if} \hspace{2mm} maxSum > 0 \hspace{2mm} \mathrm{else} \hspace{2mm} maxSum.
\end{equation}

Use [Kadane's algorithm](https://en.wikipedia.org/wiki/Maximum_subarray_problem) to find $maxSum$ and $minSum$.

```python
class Solution:
    def maxSubarraySumCircular(self, nums: List[int]) -> int:
        max_sum, cur_max = nums[0], nums[0]
        min_sum, cur_min = nums[0], nums[0]
        total = nums[0]
        
        for n in nums[1:]:
            cur_max = max(n, cur_max+n)
            max_sum = max(max_sum, cur_max)
            cur_min = min(n, cur_min+n)
            min_sum = min(cur_min, min_sum)
            total += n
            
        return max(max_sum, total - min_sum) if max_sum > 0 else max_sum
```
