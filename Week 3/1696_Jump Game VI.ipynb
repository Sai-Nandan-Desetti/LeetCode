Q: [1696. Jump Game VI](https://leetcode.com/problems/jump-game-vi/)

* Dynamic Programming (DP) problem.
* $dp[i] = \text{max score}$ achieved to reach position $i$.


* Computing $dp[i]$:
  * You can jump to a position $i$ from any of the previous $k$ positions: $i-k, i-k+1, …, i-1$.
  * To get the max score at $i$, you need to jump from a position whose score is the maximum among those positions.
\begin{equation}
  dp[i] = nums[i] + \text{max}(dp[i-k], …, dp[i-1])
\end{equation}

* The DP relation clearly requires us to compute maximum on a sliding window. Use the [239. Sliding Window Maximum](https://leetcode.com/problems/sliding-window-maximum/) problem to find it.


```python
from collections import deque

class Solution:
    def maxResult(self, nums: List[int], k: int) -> int:
        dp = [nums[0]]
        q = deque([0])        
        
        for i in range(1, len(nums)):
            dp.append(dp[q[0]] + nums[i])
            while q and dp[i] > dp[q[-1]]:
                q.pop()
            q.append(i)
            if q[0] == i-k: q.popleft()
                
        return dp[-1]
```
