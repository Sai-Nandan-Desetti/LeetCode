Q: [102. Binary Tree Level Order Traversal](https://leetcode.com/problems/binary-tree-level-order-traversal/)

**Approach 1: Iterative (BFT)**

1.1 Using a queue

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

from queue import Queue

class Solution:            
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return []
        
        q, result = Queue(), []
        q.put(root)
        
        while not q.empty():            
            level = []
            n = q.qsize()
            for _ in range(n):
                node = q.get()
                level.append(node.val)
                if node.left: q.put(node.left)
                if node.right: q.put(node.right)                
            result.append(level)            
            
        return result
````

1.2.1 Without using a queue

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:        
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        level_nodes = [root] if root else []
        node_values = []        
        while level_nodes:            
            node_values.append([node.val for node in level_nodes])
            node_pairs = [(node.left, node.right) for node in level_nodes if node]
            level_nodes = [node for pair in node_pairs for node in pair if node]       
        return node_values
````

1.2.2 Without using a queue

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:            
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return []
        
        nodes, result = [root], []        
        
        while nodes:            
            l_vals = []
            l_nodes = []
            for node in nodes:                
                l_vals.append(node.val)
                if node.left: l_nodes.append(node.left)
                if node.right: l_nodes.append(node.right)
            nodes = l_nodes
            result.append(l_vals)            
            
        return result
````

**Approach 2: Recursive (DFT)**

For the example 1 in the description, suppose that `9` had a child `5` which in turn had a child `4`, the tree is then built thus:

```
[[3]]
[[3], [9]]
[[3], [9], [5]]  # without going breadth-first to node 20, we've gone depth-first to node 5.
[[3], [9], [5], [4]]
[[3], [9, 20], [5], [4]]
[[3], [9, 20], [5, 15], [4]]
[[3], [9, 20], [5, 15, 7], [4]]
```

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:        
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return []
        
        # depth-first traversal
        def dft(root, level, result):            
            if len(result) <= level:
                result.append([])
            result[level].append(root.val)
            if root.left:
                dft(root.left, level+1, result)
            if root.right:
                dft(root.right, level+1, result)
        
        result = []
        dft(root, 0, result)
        return result
````
