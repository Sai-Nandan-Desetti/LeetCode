Q. [239. Sliding Window Maximum](https://leetcode.com/problems/sliding-window-maximum/)

* Objective: Efficiently find the maximum in a sliding window.
* Build a queue monotonically (decreasing order) so that the first element (that's why `q[0]` is appended to the `result`) is the desired max element in the window.
* A dequeue is used for this.
  * As you iterate through `nums`, if you come across an element that's (strictly) greater than the last element of the queue, you pop from the rear end as long as it is still greater.
  * Then append the element to the queue.
  * Once the window has slid past the max element (or is about to slide past, depending on the implementation), pop it from the front end of the queue.


**Approach 1:**
* Here, line 13 checks when the window is about to slide past.


```python
from collections import deque

class Solution:    
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        result, q = [], deque([])
        
        for i, n in enumerate(nums):
            while q and n > q[-1]:
                q.pop()
            q.append(n)            
            if i >= k-1:
                result.append(q[0])
                if q[0] == nums[i-k+1]: q.popleft()
                
        return result
```

**Approach 2:**
* Here, indices comprise the queue. This will turn out to be useful in [1438. Longest Continuous Subarray With Absolute Diff Less Than or Equal to Limit](https://leetcode.com/problems/longest-continuous-subarray-with-absolute-diff-less-than-or-equal-to-limit/).
* Line 12 checks once the window has just slid past.

```python
from collections import deque

class Solution:    
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        result, q = [], deque([])
        
        for i, n in enumerate(nums):
            while q and n > nums[q[-1]]:
                q.pop()
            q.append(i)
            if q[0] == i-k: q.popleft()
            if i >= k-1: result.append(nums[q[0]])
                
        return result
```
