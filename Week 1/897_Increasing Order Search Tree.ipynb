Q: [897. Increasing Order Search Tree](https://leetcode.com/problems/increasing-order-search-tree/)

* `self.build` is used for building the resultant tree. (It should be a variable whose scope and life exists outside `inorder()`.)
* By looking at lines 13 and 14, you must get an idea why we need a dummy node at the start to the right of which we construct our desired tree. (We *first* append to `self.build`'s right and then make `self.build` point to the newly added node. So, at the beginning, there must be a dummy node to the right of which we need to start building the desired tree, right?)
* This dummy node is labelled `ans` (short for "answer"). Thus, `ans.right` is the root of our desired tree.
* When building the tree, each node's left must be null. Line 12 does that.
* The recursion stops when a `node is None`.

**How the tree is built?**

Having passed a `node` to `inorder()`,
* traverse inorder along its left subtree,
* assign `node.left = None`,
* add it to the resultant tree, and
* traverse along the right subtree.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def increasingBST(self, root: TreeNode) -> TreeNode:
        def inorder(node):
            if node:
                inorder(node.left)
                node.left = None
                self.build.right = node
                self.build = node
                inorder(node.right)
        
        ans = self.build = TreeNode(None)        
        inorder(root)       
        return ans.right 
```
