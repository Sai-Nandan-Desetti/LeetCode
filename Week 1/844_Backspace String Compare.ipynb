Q: [844. Backspace String Compare](https://leetcode.com/problems/backspace-string-compare/)

* Starting from the back it's easy to see which characters are not going to be backspaced.
* Check for equality of such characters from the two strings.
* If any pair does not turn up equal, immediately return False.
* Using a generator is a nice to way to parallely generate the characters that are not going to be deleted in the two strings, respectively.

```python
class Solution:
    def backspaceCompare(self, s: str, t: str) -> bool:                
        def undeleted_chars(x):
            skip = 0
            for c in x[::-1]:
                if c == '#':
                    skip += 1
                elif skip:
                    skip -= 1
                else:
                    yield c
        try:          
            for s_c, t_c in zip(undeleted_chars(s), undeleted_chars(t), strict=True):
                if s_c != t_c:
                    return False
            return True            
        except ValueError as ve:
            return False
```
